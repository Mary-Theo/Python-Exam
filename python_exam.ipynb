```python
# Write a function called grade(score) that accepts a single number and returns "Distinction" for scores 70 and above,
#  "Pass" for scores between 50 and 69 (inclusive), or "Fail" for scores below 50.
#  Call your function with 85, 60, and 30 and print each result

scores = [ 85, 60, 30]

def grade(score: int) -> str:
    
    
    if score >= 70:
        return "Distinction"
    elif 50 <= score <= 69:
        return "Pass"
    else: 
        return "Fail"

print(grade(85))
print(grade(60))
print(grade(30))
```

```python
# Write a function called monthly_report(sales) that accepts a list of monthly sales and prints: 
# the total sales, average sales, best month number, and worst month number. Month numbers start at 1, not 0. 
# Call it with [5000, 8200, 3100, 9700, 4400, 7600].

def monthly_report(sales: list) -> int:
    
    
    total_sales = sum(sales)
    average_sales = total_sales / len(sales)

    best_month = sales.index(max(sales)) + 1
    worst_month = sales.index(min(sales)) + 1

    print("Total sales:", total_sales)
    print("Average sales:", average_sales)
    print("Best month:", best_month)
    print("Worst month:", worst_month)

monthly_report([5000, 8200, 3100, 9700, 4400, 7600])
```

```python
# Write a function called count_in_range(data, low, high) that counts how many values in a list fall between low and high (inclusive).
#  Use it to count values between 60 and 80 in [45, 72, 88, 55, 91, 60, 38, 79] and print the result.

def count_category(data: list, low: int, high: int) -> int:
    

    count = 0

    for value in data:
        if low <= value <= high:
            count += 1

    return count

result = count_category([45, 72, 88, 55, 91, 60, 38, 79], 60, 80)
print(result)
```
